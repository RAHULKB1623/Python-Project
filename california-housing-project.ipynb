## 6. Write a programming construct (create a user defined function) to calculate the median value of the data set wherever required.

```python
def calculate_median(data):
    """Calculates the median value of a given data set"""
    sorted_data = sorted(data)
    n = len(sorted_data)
    if n % 2 == 0:
        # if even number of values, take average of middle two values
        median = (sorted_data[n//2 - 1] + sorted_data[n//2]) / 2
    else:
        # if odd number of values, take middle value
        median = sorted_data[n//2]
    return median
```

## 10. Please create a new column named total_bedroom_size. If the total bedrooms is 10 or less, it should be quoted as small. If the total bedrooms is 11 or more but less than 1000, it should be medium, otherwise it should be considered large.

```python


# define function to assign bedroom size based on total_bedrooms
def assign_bedroom_size(total_bedrooms):
    if total_bedrooms <= 10:
        return 'small'
    elif total_bedrooms < 1000:
        return 'medium'
    else:
        return 'large'

# apply function to create new column total_bedroom_size
df['total_bedroom_size'] = df['total_bedrooms'].apply(assign_bedroom_size)

# save the updated data set to a CSV file
df.to_csv('data_with_bedroom_size.csv', index=False)
```
